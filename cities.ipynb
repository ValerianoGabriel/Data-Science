```python
import pandas as pd
import re
```

```python
# Reading the archive
sweethome = pd.read_csv('aus_real_estate.csv')
sweethome
```

```python
# searching for houses in Perth under $300k and built from 2015 onwards
sweethome.loc[(sweethome['City'] == 'Perth') & (sweethome['Year_Built'] > 2015) & (sweethome['Type'] == 'House') & (sweethome['Price'] < 300000)]
```

```python
#Finding errors in the state (Perth is in Western Australia - WA)
sweethome.loc[(sweethome['City'] == 'Perth') & (sweethome['State'] != 'WA')]
```

```python
# searching for the 'Real one'. Searching all the City = Perth AND State = WA
realone = sweethome.loc[(sweethome['City'] == 'Perth') & (sweethome['State'] == 'WA')]
realone
```

```python
#Replacing 'State' values with a dictionary
city_to_state= {
    'Perth': 'WA',
    'Sydney': 'NSW',
    'Adelaide': 'SA',
    'Brisbane': 'QLD'
}
```

```python
for city, state in city_to_state.items():
    sweethome.loc[sweethome['City'] == city, 'State'] = state
```
